```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pandas as pd
```

```python
uniform_data = stats.uniform.rvs(size = 10000, loc= 0, scale=10)
```

```python
uniform_data
```

```python
data_df = pd.DataFrame(uniform_data)
data_df
```

```python
data_df.plot(kind='density', xlim=(-1, 11))
```
