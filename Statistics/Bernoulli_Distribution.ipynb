```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
import pandas as pd
```

```python
bd = bernoulli(0.75)
```

```python
X = [0, 1]
```

```python
plt.figure(figsize=(10, 10))
plt.xlim(-1, 1)
plt.bar(X, bd.pmf(X))
plt.title("Bernoulli Distribution plot with size of p = 0.75")
plt.xlabel("Value of X (0, 1)")
plt.ylabel("Probability")
plt.show()
```
